### 11. How many students got A, B, C, F?

```python
criteria1 = df1["Grade"] == "A"
criteria2 = df1["Grade"] == "B"
criteria3 = df1["Grade"] == "C"
print("Students got A :",df1[criteria1].count().mean())
print("Students got B :",df1[criteria2].count().mean())
print("Students got C :",df1[criteria3].count().mean())
print("Students got F :",df1[criteria].count().mean())
```

### 12. What are the mean scores for students who got A, B, C, F?

```python
A_Aver = df1[criteria1]
print("Average of A : ", A_Aver["Total (100)"].mean())
B_Aver = df1[criteria2]
print("Average of B : ", B_Aver["Total (100)"].mean())
C_Aver = df1[criteria3]
print("Average of C : ", C_Aver["Total (100)"].mean())
F_Aver = df1[criteria]
print("Average of F : ", F_Aver["Total (100)"].mean())
```
